```python
# A0I_1
A0I_1_lat_min, A0I_1_lat_max = 0, 10       # degrees N
A0I_1_lon_min, A0I_1_lon_max = 20, 30     # degrees E

# Filter craters within the box
AOI_1 = large_craters[
    (large_craters["LAT_CIRC_IMG"] >= A0I_1_lat_min) &
    (large_craters["LAT_CIRC_IMG"] <= A0I_1_lat_max) &
    (large_craters["LON_CIRC_IMG"] >= A0I_1_lon_min) &
    (large_craters["LON_CIRC_IMG"] <= A0I_1_lon_max)
]

num_craters_1 = len(AOI_1)

# Count them
print("Number of craters ≥ 1 km in region:", num_craters_1)
```

```python
# A0I_2
A0I_2_lat_min, A0I_2_lat_max = -10, 0       # degrees N
A0I_2_lon_min, A0I_2_lon_max = 30, 40     # degrees E

# Filter craters within the box
AOI_2 = large_craters[
    (large_craters["LAT_CIRC_IMG"] >= A0I_2_lat_min) &
    (large_craters["LAT_CIRC_IMG"] <= A0I_2_lat_max) &
    (large_craters["LON_CIRC_IMG"] >= A0I_2_lon_min) &
    (large_craters["LON_CIRC_IMG"] <= A0I_2_lon_max)
]

num_craters_2 = len(AOI_2)

# Count them
print("Number of craters ≥ 1 km in region:", num_craters_2)
```

```python
def solve_for_t(N_val, P1_val, P2_val, P3_val):
    # Define symbol
    t = sp.Symbol('t', real=True)
    
    # Define the equation: N = P1*(exp(P2*t) - 1) + P3*t
    eq = sp.Eq(N_val, P1_val * (sp.exp(P2_val * t) - 1) + P3_val * t)
    
    # Solve for t
    solutions = sp.solve(eq, t)
    
    # Evaluate solutions numerically
    numeric_solutions = [sp.N(sol) for sol in solutions]
    return numeric_solutions
```
